```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("/Users/glmarschmann/.julia/dev/DEBmicroTrait/final_manuscript/files/isolates_assimilation.csv")
df = df.loc[(df.response == "positive") | (df.response == "negative")]
df = df[df.transporter_density > 1e-8]
df.KD = df.KD*1e3
df.head()
```

```python
fig_width = 6.28                         # inches  
golden_mean = (np.sqrt(5)-1.0)/2.0      # Aesthetic ratio
fig_height = fig_width*golden_mean      # height in inches
dpi = 300.0                             # Convert inch to pt

# Margins are specified as a fraction of axes size.
# May need to adjust (especially left and bottom) to accomodate axes labels/tickmarks
leftmargin = 0.15
bottommargin = 0.2
rightmargin = 0.05
topmargin = 0.1
# These two parameters are used for adjusting spacing between subplots (when used)
wspace = 0.2
hspace = 0.2

fig_size =  [fig_width,fig_height]
params = {'backend': 'svg',
          'axes.labelsize': 8,
          'axes.titlesize': 10,
          'legend.fontsize': 8,
          'xtick.labelsize': 8,
          'ytick.labelsize': 8,
          'figure.figsize': fig_size}
```

```python
plt.rcParams.update(params)
plt.figure(1)
plt.clf()
plt.axes([leftmargin,
            bottommargin,
            1.0 - rightmargin-leftmargin,
            1.0 - topmargin-bottommargin])
```

```python
ax = plt.gca()
ax = sns.stripplot(x="ontology", y="Vmax", hue="response", palette =["k", 'k'], data=df, dodge=True, size=1)
ax = sns.boxplot(x="ontology", y="Vmax", hue="response", data=df, palette =["#007100", '#bd5e8b'], showfliers=False)
ax.set(yscale="log")
ax.set(ylabel='', xlabel='')
plt.savefig("/Users/glmarschmann/.julia/dev/DEBmicroTrait/final_manuscript/plots/uptake_boxplot_Vmax.svg", format="svg")
```

```python
plt.rcParams.update(params)
plt.figure(1)
plt.clf()
plt.axes([leftmargin,
            bottommargin,
            1.0 - rightmargin-leftmargin,
            1.0 - topmargin-bottommargin])
```

```python
ax = plt.gca()
ax = sns.stripplot(x="ontology", y="KD", hue="response", palette =["k", 'k'], data=df, dodge=True, size=1)
ax = sns.boxplot(x="ontology", y="KD", hue="response", data=df, palette =["#007100", '#bd5e8b'], showfliers=False)
ax.set(yscale="log")
ax.set(ylabel='', xlabel='')
ax.set_ylim([5e-1, 5e2])
plt.savefig("/Users/glmarschmann/.julia/dev/DEBmicroTrait/final_manuscript/plots/uptake_boxplot_KD.svg", format="svg")
```

```python
plt.rcParams.update(params)
plt.figure(1)
plt.clf()
plt.axes([leftmargin,
            bottommargin,
            1.0 - rightmargin-leftmargin,
            1.0 - topmargin-bottommargin])
```

```python
ax = plt.gca()
ax = sns.stripplot(x="ontology", y="affinity", hue="response", palette =["k", 'k'], data=df, dodge=True, size=1)
ax = sns.boxplot(x="ontology", y="affinity", hue="response", data=df, palette =["#007100", '#bd5e8b'], showfliers=False)
ax.set(yscale="log")
ax.set(ylabel='', xlabel='')
plt.savefig("/Users/glmarschmann/.julia/dev/DEBmicroTrait/final_manuscript/plots/uptake_boxplot_affinity.svg", format="svg")
```
