```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("/Users/glmarschmann/.julia/dev/DEBmicroTrait/files/isolates_assimilation.csv")
df = df.loc[(df.response == "positive") | (df.response == "negative")]
df = df[df.transporter_density > 1e-8]
df.head()
```

```python
fig_width = 6.28                         # inches  
golden_mean = (np.sqrt(5)-1.0)/2.0      # Aesthetic ratio
fig_height = fig_width*golden_mean      # height in inches
dpi = 300.0                             # Convert inch to pt

# Margins are specified as a fraction of axes size.
# May need to adjust (especially left and bottom) to accomodate axes labels/tickmarks
leftmargin = 0.15
bottommargin = 0.2
rightmargin = 0.05
topmargin = 0.1
# These two parameters are used for adjusting spacing between subplots (when used)
wspace = 0.2
hspace = 0.2

fig_size =  [fig_width,fig_height]
params = {'backend': 'svg',
          'axes.labelsize': 8,
          'axes.titlesize': 10,
          'legend.fontsize': 8,
          'xtick.labelsize': 8,
          'ytick.labelsize': 8,
          'figure.figsize': fig_size,
         'text.usetex': True}
```

```python
plt.rcParams.update(params)
plt.figure(1)
plt.clf()
plt.axes([leftmargin,
            bottommargin,
            1.0 - rightmargin-leftmargin,
            1.0 - topmargin-bottommargin])
```

```python
from matplotlib.lines import Line2D
ax = plt.gca()
ax = sns.regplot(x="KD", y="Vmax", data=df,lowess=True, line_kws={"color":"b","alpha":1.0,"lw":1.5}, scatter_kws={"color":"none", "edgecolors":"k"}, ax=ax, truncate=True);
plt.xscale('log',basex=10) 
plt.yscale('log',basey=10) 
plt.minorticks_off()
ax.set_ylim([1e-4, 1e1])
#ax.set_yticklabels([1e-4, 1e-3, 1e-2, 1e-1, 1e0])
ax.set_xlim([8e-4, 1e0])
plt.xlabel("Half-saturation constant []")
plt.ylabel("Max. specific uptake rate []")
legend_elements = [Line2D([0], [0], marker='o', color='w', label='Estimated data \n(83 metabolites x 39 isolates)',
                          markerfacecolor='w',markeredgecolor="k", markersize=6),
                  Line2D([0], [0], color='b', lw=1.5, label='Nonlinear fit')]
ax.legend(handles=legend_elements, loc='best')
#plt.savefig("/Users/glmarschmann/.julia/dev/DEBmicroTrait/plots/Vmax_K_tradeoff.svg", format="svg")
```
