```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv("/Users/glmarschmann/.julia/dev/DEBmicroTrait/files/isolates2traits.csv")
```

```python
df["z_sugars_norm"] = df.z_sugars/(df.Genome_size/1e6)
df["z_organic_acids_norm"] = df.z_organic_acids/(df.Genome_size/1e6)
df["z_fatty_acids_norm"] = df.z_fatty_acids/(df.Genome_size/1e6)
df["z_auxins_norm"] = df.z_auxins/(df.Genome_size/1e6)
df["z_nucleotides_norm"] = df.z_nucleotides/(df.Genome_size/1e6)
df["z_amino_acids_norm"] = df.z_amino_acids/(df.Genome_size/1e6)
df["z_hydrolases_norm"] = df.z_hydrolases/(df.Genome_size/1e6)
```

```python
df_pn = df.loc[(df.Rhizosphere_response == "positive") | (df.Rhizosphere_response == "negative")]
```

```python
fig_height = 6.28/1.8                         # inches  
golden_mean = (np.sqrt(5)-1.0)/2.0      # Aesthetic ratio
fig_width = fig_height*golden_mean      # height in inches
dpi = 300.0                             # Convert inch to pt

# Margins are specified as a fraction of axes size.
# May need to adjust (especially left and bottom) to accomodate axes labels/tickmarks
leftmargin = 0.15
bottommargin = 0.2
rightmargin = 0.05
topmargin = 0.1
# These two parameters are used for adjusting spacing between subplots (when used)
wspace = 0.2
hspace = 0.2

fig_size =  [fig_width,fig_height]
params = {'backend': 'svg',
          'axes.labelsize': 8,
          'axes.titlesize': 10,
          'legend.fontsize': 8,
          'xtick.labelsize': 8,
          'ytick.labelsize': 8,
          'figure.figsize': fig_size}
```

```python
plt.rcParams.update(params)
plt.figure(1)
plt.clf()
plt.axes([leftmargin,
            bottommargin,
            1.0 - rightmargin-leftmargin,
            1.0 - topmargin-bottommargin])
```
