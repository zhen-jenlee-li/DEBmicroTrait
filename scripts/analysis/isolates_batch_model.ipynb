```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import h5py
```

```python
df = pd.read_csv("/Users/glmarschmann/.julia/dev/DEBmicroTrait/files/isolates_batch_model_BGE.csv")
df = df.loc[(df.response == "positive") | (df.response == "negative")]
df = df[df.BGE > 8e-2]
```

```python
max(df.BGE)
```

```python
fig_width = 6.28                         # inches  
golden_mean = (np.sqrt(5)-1.0)/2.0      # Aesthetic ratio
fig_height = fig_width*golden_mean      # height in inches
dpi = 300.0                             # Convert inch to pt

# Margins are specified as a fraction of axes size.
# May need to adjust (especially left and bottom) to accomodate axes labels/tickmarks
leftmargin = 0.15
bottommargin = 0.2
rightmargin = 0.05
topmargin = 0.1
# These two parameters are used for adjusting spacing between subplots (when used)
wspace = 0.2
hspace = 0.2

fig_size =  [fig_width,fig_height]
params = {'backend': 'svg',
          'axes.labelsize': 10,
          'axes.titlesize': 10,
          'legend.fontsize': 10,
          'xtick.labelsize': 10,
          'ytick.labelsize': 10,
          'figure.figsize': fig_size,
          'text.usetex': True}
```

```python
plt.rcParams.update(params)
plt.figure(1)
plt.clf()
plt.axes([leftmargin,
            bottommargin,
            1.0 - rightmargin-leftmargin,
            1.0 - topmargin-bottommargin])
```

```python
ax = plt.gca()
ax = sns.stripplot(x="ontology", y="BGE", hue="response", palette =["k", 'k'], data=df, dodge=True, size=1)
ax = sns.boxplot(x="ontology", y="BGE", hue="response", data=df, palette =["#007100", '#bd5e8b'], showfliers=False,  whis=[5, 95])
ax.set(yscale="log")
ax.set(ylabel='Carbon use efficiency [-]', xlabel='')
ax.set_ylim([8e-2, 1e0])
plt.savefig("/Users/glmarschmann/.julia/dev/DEBmicroTrait/plots/isolates_batch_model_bge_pn.svg", format="svg")
```
